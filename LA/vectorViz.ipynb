```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
# 2d V
# 2 inp
v2 = [4.35, -5.75]

# 3d V
# 3 inp [l, d, h]
v3 = [4.35, -4.35, -5.75]

# row => column
v3t = np.transpose(v3)

# plot
plt.plot([0, v2[0]], [0, v2[1]])
plt.axis('equal')
# map size
plt.plot([-7, 7], [0,0], 'k--')
plt.plot([0, 0], [-7, 7], 'k--')
plt.grid()
plt.axis((-7, 7, -7, 7))
plt.show()

# PLOT 3D
fig = plt.figure(figsize=plt.figaspect(1))
ax = plt.axes(None, projection='3d')
ax.plot([0, v3[0]], [0, v3[1]], [0, v3[2]], linewidth=3)

# Graphical enhancement
ax.plot([0, 0], [0, 0], [-7, 7], 'k--')
ax.plot([0,0], [-7, 7], [0,0], 'k--')
ax.plot([-7, 7], [0,0], [0,0], 'k--')
plt.show()
```
