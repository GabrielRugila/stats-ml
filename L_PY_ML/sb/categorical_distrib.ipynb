```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df
```

```python
sns.boxplot(df, y='math score', x='test preparation course')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(df, y='math score', x='parental level of education')
```

```python
df.groupby('test preparation course').mean()[['math score', 'reading score', 'writing score']].transpose()
```

```python
avg_score = df.groupby(['parental level of education','test preparation course']).mean()[['math score', 'reading score', 'writing score']].transpose()
avg_score
```

above and below represent the same data observation, just through different methods

```python
plt.figure(figsize=(10.5, 6))
sns.boxplot(df, y='math score', x='parental level of education', hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05, 1))
```

```python
plt.figure(figsize=(10.5, 6))
sns.boxplot(df, y='writing score', x='parental level of education', hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05, 1))
```
