```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Scatter Plots => relation between 2 continuous features
Ex. age, height, salary, temperature, prices, money

```python
df = pd.read_csv('dm_office_sales.csv')
df.head(10)
```

```python
fig, axes = plt.subplots(3, 1, figsize=(12, 14), dpi=120)
sns.scatterplot(ax=axes[0], x='salary', y='sales', data=df, hue='work experience', size='salary', alpha=0.6)
sns.scatterplot(ax=axes[1], x='salary', y='sales', data=df, hue='training level', size='salary', alpha=0.6)
sns.scatterplot(ax=axes[2], x='salary', y='sales', data=df, hue='level of education', palette='Dark2', size='salary', alpha=0.6)
plt.tight_layout()
plt.savefig('scatter.png')
#sns.scatterplot(x='salary', y='sales', data=df, hue='work experience')
```
