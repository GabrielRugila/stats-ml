```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
sns.catplot(df, x='gender', y='math score', kind='box', col='test preparation course', row='lunch')
```

```python
scores = sns.PairGrid(df)
scores = scores.map_upper(sns.scatterplot)
scores = scores.map_diag(sns.kdeplot)
scores = scores.map_lower(sns.histplot)
```
