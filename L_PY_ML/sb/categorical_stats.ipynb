```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(df, x='division', order=df['division'].value_counts().index)
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(df, x='division', hue='level of education')
plt.legend(bbox_to_anchor=(1.25, 1))
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(df, x='level of education')
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(df, x='level of education', hue='division', palette='Set2')
```

```python
plt.figure(figsize=(10, 4))
sns.barplot(df, x='level of education', y='salary', estimator=np.mean, errorbar='sd')
```
