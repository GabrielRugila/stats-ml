```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 11, 10)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x*x, label='x * x')
ax.plot(x, x*2, label='x * 2')

ax.legend(loc=0)
```
