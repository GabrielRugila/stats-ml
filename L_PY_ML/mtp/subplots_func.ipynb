```python
import numpy as np
import matplotlib.pyplot as plt
```

plt.subplots() => tuples
fig, axes => plt.subplots(nrows = x, ncols = y)
fig -> entire figure canvas
axes -> numpy array holding each of the axes according to position in the canvas

```python
a = np.linspace(0, 10, 10)
b = a**2.5
```

```python
x = np.arange(0, 10)
y = 3 * x + 1
```

```python
c = b + y
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0, 0].plot(a,b)
axes[1, 0].plot(x,y)
axes[0, 1].plot(x,c)
axes[1, 1].plot(a,c)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)
for ax in axes:
    ax.plot(x, y)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(5, 3), dpi=150)
axes[0, 0].plot(a,c)
axes[1, 0].plot(x,c)
axes[0, 1].plot(b,c)
axes[1, 1].plot(y,c)

axes[0, 0].set_title('AC')
axes[1, 0].set_title('XC')
axes[0, 1].set_title('BC')
axes[1, 1].set_title('YC')

fig.suptitle('By C', fontsize=16)
fig.subplots_adjust(hspace=0.7, wspace=0.5)

fig.savefig('subplots_c.png', bbox_inches='tight')
```
