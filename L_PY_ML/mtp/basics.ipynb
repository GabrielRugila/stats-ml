```python
x = np.arange(0, 10)
y = x ** 3.14159265359
# multiple axes
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 0.6, 0.6])
axes1.plot(x, y)
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('Power of π')

axes2 = fig.add_axes([0.195, 0.2, 0.2, 0.2])
axes2.plot(x, y)
axes2.set_xlim(3, 5.75)
axes2.set_ylim(0, 250)
axes2.set_title('Zoomed in')

plt.show()
```
