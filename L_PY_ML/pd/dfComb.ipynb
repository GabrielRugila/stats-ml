```python
# merge / join
# .merge() => inner, left, right
# on (column) => unique identifier & must be present on both tables
# merge => venn diagram
# how="inner" => records that match in both tables
```

```python
r1 = {'reg id':[1,2,3,4], 'name':['Andrew', 'Bob', 'Chelsy', 'Dave']}
l1 = {'log id':[1,2,3,4], 'name':['Xavier', 'Mark', 'Bob', 'John']}

dfr1 = pd.DataFrame(r1)
dfl1 = pd.DataFrame(l1)
```

```python
dfr1
```

```python
dfl1
```

```python
pd.merge(dfr1,dfl1,how="inner", on='name')
```

```python
pd.merge(dfr1,dfl1,how="right", on='name')
```

```python
help(pd.merge)
```

```python
pd.merge(dfl1,dfr1,how="left", on='name')
```

```python
pd.merge(dfr1,dfl1,how="left", on='name')
```

```python
pd.merge(dfr1,dfl1,how="outer", on='name')
```

```python
pd.merge(dfr1,dfl1,how="outer", on='name', indicator=True)
```
