```python
def last_n(num, cutN):
    return str(num)[-cutN:]

def last_four(num):
    return str(num)[-4:]
```

```python
def yelp(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 15:
        return '$$'
    elif price >= 15 and price < 25:
        return '$$$'
    elif price >= 25 and price < 30:
        return '$$$$'
    else:
        return '$$$$$'
```

```python
df['total_bill'].apply(yelp)
```

```python
df['Yelp'] = df['total_bill'].apply(yelp)
```

```python
# Lambda Expression

# lambda num: num*2
# num = just rand var name

df['total_bill'].apply(lambda num: num*2)
```

```python
def quality(total_bill, tip):
    if tip/total_bill > 0.15:
        return "Generous"
    else:
        return "Other"
```

```python
df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
```

```python
np.vectorize(quality)(df['total_bill'], df['tip'])
```

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
df
```
